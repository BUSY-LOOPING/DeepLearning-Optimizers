```python
import numpy as np

# define the saddle function J(theta1, theta2)
def J(theta1, theta2):
    return theta1**2 - theta2**2

# define the gradient of the saddle function
def grad_J(theta1, theta2):
    return np.array([2*theta1, -2*theta2])

# set the initial parameters
theta = np.array([3, 0])

# set the learning rate and number of iterations
alpha = 0.1
num_iterations = 100
theta1_hist = []
theta2_hist = []

# run stochastic gradient descent
for i in range(num_iterations):
    # randomly choose a data point
    idx = np.random.randint(0, 2)
    if idx == 0:
        data = np.array([1, 0])
    else:
        data = np.array([0, 1])
    # calculate the gradient at the current parameters for the chosen data point
    gradient = grad_J(theta[0]*data[0], theta[1]*data[1])
    
    # update the parameters
    theta = theta - alpha*gradient
    theta1_hist.append(theta[0])
    theta2_hist.append(theta[1])

    # print the current iteration and parameters
    print(f"Iteration {i+1}: theta1 = {theta[0]}, theta2 = {theta[1]}, J = {J(theta[0], theta[1])}")
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(theta1_hist, label = 'theta1')
plt.plot(theta2_hist, label = 'theta2')

plt.legend()
plt.show()
```

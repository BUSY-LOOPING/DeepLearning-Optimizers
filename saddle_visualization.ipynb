```python
x = np.random.randn(1000)
y= np.random.randn(1000)
```

```python
def fun(x, y) :
    return (x ** 2) - (y ** 2)
```

```python
z = fun(x, y)
```

```python
#3d plot
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot_trisurf(x, y, z, linewidth = 0.2, antialiased = True, cmap = 'jet')
plt.show()
```

```python
#3d plot
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
x1 = y1 = np.arange(-3.0, 3.0, 0.05)
X1, Y1 = np.meshgrid(x1, y1)
z1 = fun(X1, Y1)
ax.plot_surface(X1, Y1, z1, linewidth = 0.2, antialiased = True, cmap = 'jet')
plt.show()
```
